```python
import torch
from tqdm import tqdm
import networkx as nx
import numpy as np
```

```python
G = nx.read_adjlist('./SCHOLAT Link Prediction/train.csv',nodetype=int,delimiter=',')
```

```python
a = []
S = G.subgraph(max(nx.connected_components(G),key=len))
for node in dict(nx.shortest_path_length(S)).values():
    for v in node.values():
        a.append(v)

res = np.mean(a)
res
```
